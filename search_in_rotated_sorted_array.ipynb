3Ô∏è‚É£ Why Binary Search Is Faster?

Binary Search uses Divide and Conquer.

Idea:

Instead of checking every element, it:
	‚Ä¢	Looks at the middle
	‚Ä¢	Discards half the data every step

üìâ Data size reduces like:

	n ‚Üí n/2 ‚Üí n/4 ‚Üí n/8 ‚Üí ...

That‚Äôs why complexity is O(log n).

9Ô∏è‚É£ Basic Binary Search Code (Reference)

def binary_search(arr, target):
    low, high = 0, len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1

1Ô∏è‚É£2Ô∏è‚É£ Common Mistakes to Avoid ‚ùå

Question:

Suppose an array of length n sorted in ascending order is rotated between 1 and n times. For example, the array nums = [0,1,2,4,5,6,7] might become:

[4,5,6,7,0,1,2] if it was rotated 4 times.
[0,1,2,4,5,6,7] if it was rotated 7 times.
Notice that rotating an array [a[0], a[1], a[2], ..., a[n-1]] 1 time results in the array [a[n-1], a[0], a[1], a[2], ..., a[n-2]].

Given the sorted rotated array nums of unique elements, return the minimum element of this array.

You must write an algorithm that runs in O(log n) time.



Example 1:

Input: nums = [3,4,5,1,2]
Output: 1
Explanation: The original array was [1,2,3,4,5] rotated 3 times.
Example 2:

Input: nums = [4,5,6,7,0,1,2]
Output: 0
Explanation: The original array was [0,1,2,4,5,6,7] and it was rotated 4 times.
Example 3:

Input: nums = [11,13,15,17]
Output: 11
Explanation: The original array was [11,13,15,17] and it was rotated 4 times.


Constraints:

n == nums.length
1 <= n <= 5000
-5000 <= nums[i] <= 5000
All the integers of nums are unique.
nums is sorted and rotated between 1 and n times.

‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª‚∏ª

Answer:

This is the classic ‚ÄúFind Minimum in Rotated Sorted Array‚Äù problem.
Because the array is sorted, rotated, and contains unique elements, we can use binary search to achieve O(log n) time.

‚∏ª

Key Insight
	‚Ä¢	A rotated sorted array consists of two sorted parts.
	‚Ä¢	The minimum element is the pivot where rotation happened.
	‚Ä¢	Compare the middle element with the rightmost element to decide which half contains the minimum.

‚∏ª

Binary Search Strategy

Let l = 0, r = n - 1.

While l < r:
	‚Ä¢	Compute mid = (l + r) // 2
	‚Ä¢	If nums[mid] > nums[r]
	‚Ä¢	The minimum lies to the right of mid
	‚Ä¢	Set l = mid + 1
	‚Ä¢	Else
	‚Ä¢	The minimum lies at mid or to the left
	‚Ä¢	Set r = mid

When the loop ends, l == r and points to the minimum element.

‚∏ª

Why This Works
	‚Ä¢	If nums[mid] > nums[r], the rotation point is after mid
	‚Ä¢	If nums[mid] <= nums[r], the right half is sorted, so the minimum is not beyond mid
	‚Ä¢	Each step halves the search space ‚Üí O(log n)

‚∏ª

Python Implementation

def findMin(nums):
    l, r = 0, len(nums) - 1

    while l < r:
        mid = (l + r) // 2

        if nums[mid] > nums[r]:
            l = mid + 1
        else:
            r = mid

    return nums[l]

```python
class Solution:
  def FindMin(self,nums):
    l, r = 0, len(nums) - 1
    while l < r:
      mid = (l+r) // 2
      if nums[mid]<nums[r]:
        l = mid + 1
      else:
        r = mid
    return nums[l]
```

```python
class Solution:
  def FindMin(self, nums):
    l, r = 0, len(nums) - 1
    while l < r:
      mid = (l + r) // 2
      # If nums[mid] > nums[r], the minimum must be in the right half (mid+1 to r)
      # because the left part (l to mid) is sorted and greater than nums[r]
      if nums[mid] > nums[r]:
        l = mid + 1
      # If nums[mid] <= nums[r], the minimum is in the left half (l to mid)
      # or is nums[mid] itself. So we move r to mid.
      else:
        r = mid
    # When the loop finishes, l == r, and this index points to the minimum element.
    return nums[l]

# Test cases
sol = Solution()

# Example 1: [3,4,5,1,2] -> 1
nums1 = [3,4,5,1,2]
expected1 = 1
result1 = sol.FindMin(nums1)
print(f"Input: {nums1}, Output: {result1}, Expected: {expected1}")
assert result1 == expected1, f"Test Case 1 Failed: Expected {expected1}, got {result1}"

# Example 2: [4,5,6,7,0,1,2] -> 0
nums2 = [4,5,6,7,0,1,2]
expected2 = 0
result2 = sol.FindMin(nums2)
print(f"Input: {nums2}, Output: {result2}, Expected: {expected2}")
assert result2 == expected2, f"Test Case 2 Failed: Expected {expected2}, got {result2}"

# Example 3: [11,13,15,17] -> 11 (already sorted)
nums3 = [11,13,15,17]
expected3 = 11
result3 = sol.FindMin(nums3)
print(f"Input: {nums3}, Output: {result3}, Expected: {expected3}")
assert result3 == expected3, f"Test Case 3 Failed: Expected {expected3}, got {result3}"

# Additional test cases

# Single element array
nums4 = [1]
expected4 = 1
result4 = sol.FindMin(nums4)
print(f"Input: {nums4}, Output: {result4}, Expected: {expected4}")
assert result4 == expected4, f"Test Case 4 Failed: Expected {expected4}, got {result4}"

# Two elements, rotated
nums5 = [2,1]
expected5 = 1
result5 = sol.FindMin(nums5)
print(f"Input: {nums5}, Output: {result5}, Expected: {expected5}")
assert result5 == expected5, f"Test Case 5 Failed: Expected {expected5}, got {result5}"

# Two elements, sorted
nums6 = [1,2]
expected6 = 1
result6 = sol.FindMin(nums6)
print(f"Input: {nums6}, Output: {result6}, Expected: {expected6}")
assert result6 == expected6, f"Test Case 6 Failed: Expected {expected6}, got {result6}"

# More complex rotation
nums7 = [6,7,0,1,2,3,4,5]
expected7 = 0
result7 = sol.FindMin(nums7)
print(f"Input: {nums7}, Output: {result7}, Expected: {expected7}")
assert result7 == expected7, f"Test Case 7 Failed: Expected {expected7}, got {result7}"

# Array with negative numbers
nums8 = [-5,-4,-3,-2,-1,0,1,2,3,-6]
expected8 = -6
result8 = sol.FindMin(nums8)
print(f"Input: {nums8}, Output: {result8}, Expected: {expected8}")
assert result8 == expected8, f"Test Case 8 Failed: Expected {expected8}, got {result8}"

# Array rotated such that minimum is at the last index initially
nums9 = [3,1,2]
expected9 = 1
result9 = sol.FindMin(nums9)
print(f"Input: {nums9}, Output: {result9}, Expected: {expected9}")
assert result9 == expected9, f"Test Case 9 Failed: Expected {expected9}, got {result9}"

print("All test cases passed!")
```
